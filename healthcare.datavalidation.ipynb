```python
#import matplotlib to visualize dataset
import matplotlib.pyplot as plt
```

```python
#import seaborn library
import seaborn as sns
```

```python
# Get the counts for heart disease cases
heart_disease_counts = df['HeartDiseaseorAttack'].value_counts()

# Define labels
labels = ['No Heart Disease', 'Heart Disease']
colors = ['lightgreen', 'lightcoral']
# pie chart
plt.figure(figsize=(6, 6))
plt.pie(heart_disease_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Heart Disease Prevalence')
plt.axis('equal')  # Equal aspect ratio to ensure it's a perfect circle
plt.show()
```

```python
#Distribution of hear disease cases
sns.countplot(x='HeartDiseaseorAttack', data = df)
plt.title('Distribution of heart disease cases')
plt.xlabel('Heart disease(0 = NO,1 = YES)')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
#Distribution by gender as count plot
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex', data=df, hue='Sex', palette='pastel', legend=False)
plt.title('Distribution by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Count')
plt.show()
```

```python
#Distribution of gender represented in pie chart
df['Sex'].value_counts().plot.pie(
    labels = ['Female','Male'],
    autopct= '%1.1f%%',
    colors = ['lightpink', 'lightblue'],
    startangle = 90)
plt.title('Sex Distribution')
plt.ylabel('')
plt.show()
    
```

```python
#visualization of count different age groups
age_map = {
    1:'18-24',2:'25-29',3:'30-34',4:'35-39',
    5:'40-44',6:'45-49',7:'50-54',8:'55-59',
    9:'60-64',10:'65-69',11:'70-74',12:'75-79',
    13:'80+'}
df["AgeGroup"] = df['Age'].map(age_map)

#Plot the distribution using countplot
plt.figure(figsize = (12,6))
sns.countplot(x = "AgeGroup",data = df,order = age_map.values(),hue='Age',palette = 'pastel',legend = False)
plt.title("Distribution of Age Groups")
plt.xlabel('Age Group')
plt.ylabel('count')
plt.tight_layout()
plt.show()
           
```

```python
#Histogram plot of Physical health for past 30 days
plt.figure(figsize = (10,5))
sns.histplot(df["PhysHlth"],bins=31,kde = True,color = 'darkred')
plt.title("Distribution of physical health days(past 30 days")
plt.xlabel('Number of days with Poor Physcial Health')
plt.ylabel('frequency')
plt.tight_layout()
plt.show()
          
```

```python
#Histogram plot of Mental health for past 30 days
plt.figure(figsize = (10,5))
sns.histplot(df["MentHlth"],bins = 31,kde = True,color = "orchid")
plt.title("Distribution of Mental Health from past 30 days")
plt.xlabel("Number of days with Poor Mental Health")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# List of health condition columns
health_conditions = ['HighBP', 'HighChol', 'Stroke', 'Diabetes', 'DiffWalk']

# Bar plots for each condition
fig, axes = plt.subplots(3, 2, figsize=(14, 12))
axes = axes.flatten()

for i, condition in enumerate(health_conditions):
    # Count 0s and 1s
    counts = df[condition].value_counts().sort_index()
    
    # Labels (0 = No, 1 = Yes)
    labels = ['No', 'Yes'] if len(counts) == 2 else counts.index.astype(str)
    
    # Bar plot
    axes[i].bar(labels, counts.values, color=['lightgreen', 'salmon'][:len(labels)])
    axes[i].set_title(f'Prevalence of {condition}')
    axes[i].set_ylabel('Count')

# Remove last subplot if the number is odd
if len(health_conditions) % 2 != 0:
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```
